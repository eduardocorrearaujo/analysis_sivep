### Região metropolitana do PR

```python
df = pd.read_csv('dados/proj_pop_PR.csv')
del df['Unnamed: 0'] 
df
```

```python
cidades = [4100202, # Adrianópolis
           4100301, # Agudos do Sul
           4100400, # Almirante Tamandaré 
           4101804, # Araucária 
           4102307, # Balsa Nova 
           4103107, # Bocaiúva do Sul 
           4104006, # Campina Grande do Sul
           4104105, # Campo do Tenente 
           4104204, # Campo Largo 
           4104253, # Campo Magro
           4105201, # Cerro Azul 
           4105805, # Colombo
           4106209, # Contenda
           4106902, # Curitiba
           4128633, # Doutor Ulysses 
           4107652, # Fazenda Rio Grande
           4111258, # Itaperuçu
           4113205, # Lapa
           4114302, # Mandirituba 
           4119103, # Piên
           4119152, # Pinhais
           4119509, # Piraquara  
           4120804, # Quatro Barras
           4121208, # Quitandinha 
           4122206, # Rio Branco do Sul
           4122305, # Rio Negro 
           4125506, # São José dos Pinhais 
           4127601, # Tijucas do Sul 
           4127882]  # Tunas do Sul
```

```python
df.columns
```

```python
df = df[df.Armenor.isin(cidades)]
```

```python
df_pop = pd.DataFrame()
df_aux = pd.DataFrame()
for i in [2019, 2020, 2021, 2022]:
    df_aux['local'] = ['02RS METROPOLITANA'] 
    
    df_aux['ano'] = i

    df_aux['[0,20)'] = [df.loc[df.Ano == i]['0 a 5'].sum() + df.loc[df.Ano == i]['5 a 10'].sum() + df.loc[df.Ano == i]['10 a 15'].sum() + df.loc[df.Ano == i]['15 a 20'].sum()]

    df_aux['[20,60)'] = [df.loc[df.Ano == i]['20 a 25'].sum() + df.loc[df.Ano == i]['25 a 30'].sum() + df.loc[df.Ano == i]['30 a 35'].sum() + df.loc[df.Ano == i]['35 a 40'].sum() + df.loc[df.Ano == i]['40 a 45'].sum() + df.loc[df.Ano == i]['45 a 50'].sum() + df.loc[df.Ano == i]['50 a 55'].sum() +  df.loc[df.Ano == i]['55 a 60'].sum()]

    df_aux['[60,80)'] = [df.loc[df.Ano == i]['60 a 65'].sum() + df.loc[df.Ano == i]['65 a 70'].sum() + df.loc[df.Ano == i]['70 a 75'].sum() + df.loc[df.Ano == i]['75 a 80'].sum()]

    df_aux['80+'] =  [df.loc[df.Ano == i]['80 a 85'].sum() + df.loc[df.Ano == i]['85 a 90'].sum() + df.loc[df.Ano == i]['90+'].sum()] 
    
    df_pop = pd.concat([df_pop, df_aux])
df_pop
```

```python
df_pop.to_csv('dados/pop_reg_metr_cwb.csv') 
```
