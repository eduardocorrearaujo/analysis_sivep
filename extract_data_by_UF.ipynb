Esse projeto fornece projeções da população estratificados por faixa etária e sexo para os municípios do Brasil entre 2010 e 2030. 

```python
df = pd.read_excel('dados/ProjMunic.xlsx')
df
```

## Filtrando os dados do SIVEP para o estados do paraná:

```python
%%time
chunks = pd.read_csv('dados/INFLUD_ALL.csv', chunksize = 10000, low_memory = False)

df_pr = pd.DataFrame()

for df in chunks: 
    
    df_aux = df.loc[df.SG_UF_NOT == 'PR']
    
    df_pr = pd.concat([df_pr, df_aux])

df_pr
```

```python
del df_pr['Unnamed: 0']

df_pr.to_csv('dados/INFLUD_PR.csv')
```
